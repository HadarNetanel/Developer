```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
led = np.zeros((6,50),dtype=bool)
with open('./secret_codes.txt') as f:
    for line in f:
        op, p1, p2 = line.split()
        p1, p2 = int(p1) ,int(p2)
        if op == "+":
            led[:p2,:p1] = True
        if op == ">":
            led[p1] = np.roll(led[:,p1], p2)
        if op == "v":
            led[:,p1] = np.roll(led[:,p1], p2)

plt.imshow(led)
```
